```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
image1=cv.imread('brain.png',cv.IMREAD_GRAYSCALE)
plt.imshow(image1,cmap='gray')
```

```python
kernel3 = np.ones((3, 3), np.float32) / 9
kernel5 = np.ones((5, 5), np.float32) / 25
kernel7 = np.ones((7, 7), np.float32) / 49

image3 = cv.filter2D(src=image1, ddepth=-1, kernel=kernel3)
image5 = cv.filter2D(src=image1, ddepth=-1, kernel=kernel5)
image7 = cv.filter2D(src=image1, ddepth=-1, kernel=kernel7)
plt.figure(figsize=(6, 7))

plt.subplot(2, 2, 1),plt.imshow(image1, cmap='gray'),plt.title('orginal')
plt.subplot(2, 2, 2),plt.imshow(image3, cmap='gray'),plt.title('3*3')
plt.subplot(2, 2, 3),plt.imshow(image5, cmap='gray'),plt.title('5*5')
plt.subplot(2, 2, 4),plt.imshow(image7, cmap='gray'),plt.title('7*7')

```

```python
image3 = cv.medianBlur(image1, 3)
image5 = cv.medianBlur(image1, 5)
image7 = cv.medianBlur(image1, 7)

plt.figure(figsize=(6, 7))

plt.subplot(2, 2, 1),plt.imshow(image1, cmap='gray'),plt.title('orginal')
plt.subplot(2, 2, 2),plt.imshow(image3, cmap='gray'),plt.title('3*3')
plt.subplot(2, 2, 3),plt.imshow(image5, cmap='gray'),plt.title('5*5')
plt.subplot(2, 2, 4),plt.imshow(image7, cmap='gray'),plt.title('7*7')
```
