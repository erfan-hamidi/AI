```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
im = Image.open('bridge.gif').convert('RGB')
im.save('bridge.jpg')
```

```python
image1=cv.imread('bridge.jpg',cv.IMREAD_GRAYSCALE)
plt.imshow(image1,cmap='gray')
```

```python
kernel3 = np.ones((3, 3), np.float32) / 9
kernel5 = np.ones((5, 5), np.float32) / 25
kernel7 = np.ones((7, 7), np.float32) / 49

image3 = cv.filter2D(src=image1, ddepth=-1, kernel=kernel3)
image5 = cv.filter2D(src=image1, ddepth=-1, kernel=kernel5)
image7 = cv.filter2D(src=image1, ddepth=-1, kernel=kernel7)
 
plt.subplot(1, 3, 1),plt.imshow(image3, cmap='gray'),plt.title('3*3')
plt.subplot(1, 3, 2),plt.imshow(image5, cmap='gray'),plt.title('5*5')
plt.subplot(1, 3, 3),plt.imshow(image7, cmap='gray'),plt.title('7*7')

```

```python
laplacian_kernel = np.array([[0, -1, 0],[-1, 5, -1],[0, -1, 0]])

sharpened_image = cv.filter2D(image3, -1, laplacian_kernel)
blurred_image_rescaled = cv.resize(image3, None, fx=2, fy=2, interpolation=cv.INTER_CUBIC)
sharpened_image_rescaled = cv.resize(sharpened_image, None, fx=2, fy=2, interpolation=cv.INTER_CUBIC)
```

```python
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(blurred_image_rescaled, cmap='gray')
plt.title('Blurred 3*3 Image')


plt.subplot(1, 2, 2)
plt.imshow(sharpened_image_rescaled, cmap='gray')
plt.title('Sharpened Image')
```

```python
second_sharpened_image = cv.filter2D(sharpened_image, -1, laplacian_kernel)
second_sharpened_image_rescaled = cv.resize(second_sharpened_image, None, fx=2, fy=2, interpolation=cv.INTER_CUBIC)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(blurred_image_rescaled, cmap='gray')
plt.title('Blurred Image')


plt.subplot(1, 3, 2)
plt.imshow(sharpened_image_rescaled, cmap='gray')
plt.title('1st Sharpening')


plt.subplot(1, 3, 3)
plt.imshow(second_sharpened_image_rescaled, cmap='gray')
plt.title('2nd Sharpening')
```
