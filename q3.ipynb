```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
image1=cv2.imread('Edge.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(image1,cmap='gray')
```

```python
sobel_x = cv2.Sobel(image1, cv2.CV_64F, 1, 0, ksize=3)
sobel_x = np.uint8(np.absolute(sobel_x))


sobel_y = cv2.Sobel(image1, cv2.CV_64F, 0, 1, ksize=3)
sobel_y = np.uint8(np.absolute(sobel_y))
sobel_combined = cv2.bitwise_or(sobel_x, sobel_y)
```

```python
# Display the images
plt.figure(figsize=(10, 6))

plt.subplot(2, 3, 1), plt.imshow(image1, cmap='gray')
plt.title('Original Image'), 

plt.subplot(2, 3, 2), plt.imshow(sobel_x, cmap='gray')
plt.title('Sobel X'), 

plt.subplot(2, 3, 3), plt.imshow(sobel_y, cmap='gray')
plt.title('Sobel Y'), 

plt.subplot(2, 3, 4), plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Combined'), 
```

```python
laplacian = cv2.Laplacian(image1, cv2.CV_64F)

# Calculate the magnitude of the gradient
gradient_magnitude = np.abs(laplacian)

plt.figure(figsize=(10, 6))

plt.subplot(1, 3, 1), plt.imshow(image1, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 3, 2), plt.imshow(np.uint8(np.absolute(laplacian)), cmap='gray')
plt.title('Laplacian Edge Detection')

plt.subplot(1, 3, 3), plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Gradient Magnitude')
```
